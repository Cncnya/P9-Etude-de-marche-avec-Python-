### DISPO VOLAILLES

```python
# On sélectionne les lignes avec Produit = Viande de Volailles pour notre analyse

dispo_poulet = dispo.loc[dispo['Produit'] == 'Viande de Volailles']
```

```python
# On sélectionne les colonnes utiles

dispo_poulet = dispo_poulet[['Zone', 'Élément' ,'Produit','Année', 'Unité' ,'Valeur']]
dispo_poulet.head()
```

```python
# On ramene la disponibilité par unité

dispo_poulet['Valeur'] = dispo_poulet['Valeur']*1000
dispo_poulet.head()
```

```python
dispo_poulet.info()
```

```python
# Contrôle des doublons

dispo_poulet.duplicated().sum()
```

```python
# Contrôle des valeurs nulles

dispo_poulet.isna().sum()
```

### PIVOT TABLE

```python
# On sélectionne les colonnes utiles

dispo_poulet = dispo_poulet[['Zone', 'Élément' ,'Valeur']]
dispo_poulet.head()
```

```python
poulet = dispo_poulet.pivot_table(columns='Élément', index=['Zone'], values='Valeur',aggfunc=np.sum)
poulet
```

#### Exportations

```python
poulet.loc[poulet['Exportations - Quantité'].isna()]
```

```python
# Calcul exportation 

exp = poulet['Importations - Quantité'] + poulet['Production'] + poulet['Variation de stock'] - poulet['Disponibilité intérieure']
```

```python
# On remplace les valeur NaN

poulet["Exportations - Quantité"].fillna(value = exp, inplace = True)
```
